```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
path = "D:/Users/PiotrM/Downloads/obrazki"

table = []

for filename in os.listdir(path):
    if filename.endswith(".jpg"):
        table.append(filename)
new_table = []
for filename in table:
    parts = filename.split('_',3)
    new_table.append(parts[0:3])
new_table
df = pd.DataFrame(new_table, columns=['age', 'gender', 'race'])
df = df.applymap(lambda x: pd.to_numeric(x, errors='coerce')).dropna()

# convert all values to integers
df = df.astype(int)
df
```

```python
df['age'].hist(bins=100)
plt.figure(figsize=(100, 5))
plt.show()
```

```python
df['gender'].hist()
plt.figure(figsize=(100, 5))
plt.show()
```

```python
df['race'].hist()
plt.figure(figsize=(100, 5))
plt.show()
```

```python
df['age'].plot.box()
plt.figure(figsize=(100, 5))
plt.show()
```

```python
df.plot.scatter(x='age', y='race')
plt.figure(figsize=(100, 100))

# show the plot
plt.show()
```
