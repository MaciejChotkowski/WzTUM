```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
path = "D:/Users/PiotrM/Downloads/obrazki"

table = []

for filename in os.listdir(path):
    if filename.endswith(".jpg"):
        table.append(filename)
new_table = []
for filename in table:
    parts = filename.split('_',3)
    new_table.append(parts[0:3])
new_table
df = pd.DataFrame(new_table, columns=['age', 'gender', 'race'])
df = df.applymap(lambda x: pd.to_numeric(x, errors='coerce')).dropna()

# convert all values to integers
df = df.astype(int)
df
```

```python
df['age'].hist(bins=100)
plt.figure(figsize=(100, 5))
plt.show()
```

```python
df['gender'].hist()
plt.figure(figsize=(100, 5))
plt.show()
```

```python
df['race'].hist()
plt.figure(figsize=(100, 5))
plt.show()
```

```python
df['age'].plot.box()
plt.figure(figsize=(100, 5))
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(15, 5))

# loop through each subplot and create a box plot for each group
for i, ax in enumerate(axs):
    df[df['race'] == i].boxplot(ax=ax, column=['age'], grid=False)
    ax.set_title('Race {}'.format(i))
    ax.set_xlabel('Race')
    ax.set_ylabel('age')

# remove the extra space between subplots
plt.tight_layout()

# show the plot
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# loop through each subplot and create a box plot for each group
for i, ax in enumerate(axs):
    df[df['gender'] == i].boxplot(ax=ax, column=['age'], grid=False)
    ax.set_title('Gender {}'.format(i))
    ax.set_xlabel('gender')
    ax.set_ylabel('age')

# remove the extra space between subplots
plt.tight_layout()

# show the plot
plt.show()
```
